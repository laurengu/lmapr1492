```python
from pymatgen import Structure
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
with MPRester("ZPSKgimujRNFyTRi") as m:

    # Structure pour material id
    structure = m.get_structure_by_material_id("mp-961687")

    #objet XRDCalculator
    xrd = XRDCalculator(1.54060)

    # modèle de diffraction
    modele = xrd.get_pattern(structure)
```

# Diffractogramme
Le diffractogramme permet d'étudier le comportement d'un rayon incident CuKa pour notre cristal réspectif. Ces rayons X vont créer des phénomènes d'interférences constructives ou destructives selon la direction. Sur base du modèle de diffraction, nous pouvons trouver la position de chacun des atomes du cristal.

nous considèrerons 2θ comme l'angle entre la direction d'incidence et celle d'observation

```python
#Affichage des pics du diffractogramme
xrd.show_plot(structure,two_theta_range=[0,90],annotate_peaks=False)
```

```python
theta1 = modele.x[0]
i1 = modele.y[0]
hkl1 = modele.hkls[0]

print("theta = "+str(theta1))
print("intensité = "+str(i1))
print("indices = "+str(hkl1))
```
