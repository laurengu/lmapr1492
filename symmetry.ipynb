```python
from pymatgen import Structure
from pymatgen.core.operations import SymmOp
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
from pymatgen.ext.matproj import MPRester
with MPRester("ZPSKgimujRNFyTRi") as m:

    # Structure pour material id
    structure = m.get_structure_by_material_id("mp-961687")
```

# Réflexion
La réflexion est une opération qui inverse la chiralité et dont le déterminant de l'opérateur vaut -1. On observe que l'atome garde la même coordonnée en z.


```python
SA = SpacegroupAnalyzer(structure)
new = SA.get_conventional_standard_structure() #structure conventionelle
SP = SpacegroupAnalyzer(new)
site = new.sites #sites de la structure


########Réflexion########
PM = SymmOp.from_xyz_string('y,x,z')
#Position initiale de l'atome Sn #5
Sni = site[5].frac_coords
print("coordonnées initiales de l'atome Sn #5 :", Sni)
#Position après opération de symétrie de l'atome Sn #5
Snf = PM.operate(Zri)
print("coordonnées finales de l'atome Sn #5 :", Snf)
```
